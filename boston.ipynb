```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
```

Code Loading

```python
#Add header and split column

# Column names

column_names = [

    "CRIM", "ZN", "INDUS", "CHAS", "NOX",

    "RM", "AGE", "DIS", "RAD", "TAX",

    "PTRATIO", "B", "LSTAT", "MEDV"

]

 

# Read the data file using space as the separator and without headers

data_path = "housing.csv"  # Update with the correct file path

boston = pd.read_csv(data_path, sep=r'\s+', header=None, names=column_names)

 

# Display the DataFrame

print(boston)
```

```python
#replacing words

boston.replace('HOW', np.nan, inplace=True)

boston.replace('ARE', np.nan, inplace=True)

boston.replace('?', np.nan, inplace=True)

boston.replace('U', np.nan, inplace=True)

 

# Display the DataFrame

print(boston)
```

```python
#fill NaN

boston.fillna(0, inplace=True)


# Display the DataFrame

print(boston)
```

```python
# Convert specific columns to float

columns_to_convert = ['CRIM', 'ZN'] #

#Add column names as needed # Convert the specified columns to float

boston[columns_to_convert] = boston[columns_to_convert].astype(float)
```
